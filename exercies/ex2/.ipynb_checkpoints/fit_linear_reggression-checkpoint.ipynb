```python
import numpy as np
from copy import deepcopy
import pandas as pd
from plotnine import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
_dataset = pd.read_csv("./kc_house_data.csv")[0:2000]
print(_dataset)

```

```python
def expand_bias(design_matrix):
    _shape = design_matrix.shape
    _ones = np.ones( (_shape[0], _shape[1] + 1) )
    _ones[:,:-1] = design_matrix 
    return _ones

def fit_linear_reggression(design_matrix, respone_vec, bias=False):
    if bias:
        design_matrix = expand_bias( design_matrix )
        print(design_matrix)
    
    u, s, vh = np.linalg.svd(design_matrix)
    s_dagger = deepcopy( s )
    mask = s != 0
    s_dagger[ mask ] = 1 / s [ mask ]    
    def diag_mul( _diag, _vec):
        return _diag * _vec[:_diag.shape[0]]
    t =  vh.transpose() @ diag_mul( s_dagger ,u.transpose() @ respone_vec )
    return t, s
   
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

data = _dataset.loc[:, ['price', 'yr_built', 'sqft_living']]
data = data[(data['price'] > 0)]
Y = data['price'].to_numpy()
X = data[['yr_built', 'sqft_living' ]].to_numpy()
W, S =  fit_linear_reggression(X,Y)
print(W)


ax.scatter( data['yr_built'].to_numpy(), data['sqft_living'].to_numpy(), Y )

xx, yy = np.meshgrid(range(1750,2000), range(1000,4000))
z = W[0] * xx + W[1] * yy 
ax.plot_surface(xx, yy, z, alpha=0.2)
ax.set_zlim(0.5*10**6, 0.75 * 10**6)
ax.set_ylim(0 ,4000)
# ggplot(data, aes(x='yr_built' , y='price')) +\
# geom_point()+ geom_abline(slope =W[0])  +ylim(0.5*10**6, 0.75 * 10**6)
plt.show()



```

```python
def predict( _design_matrix, _coefficients, bias = False):
    if bias:
        _design_matrix = expand_bias( _design_matrix )
        #np.append(_coefficients, 1)
    
    return _design_matrix @ _coefficients
```

```python
def mse(predicvec, responevec ):
    return np.linalg.norm(responevec - predicvec) ** 2 / responevec.shape[0]

```

```python
def load_data(_path):
    _dataset_prepoc = pd.read_csv(_path)
    target = 'price'
    prices = _dataset_prepoc[target].to_numpy()
    droped_fe = ['id' , 'date', 'zipcode', 'yr_renovated']
    categorical = ['view' , 'waterfront', 'bedrooms', 'grade']
    cat = pd.get_dummies(_dataset_prepoc[categorical].astype('category'))
    _dataset_prepoc = _dataset_prepoc.drop([target] + categorical + droped_fe, axis=1).to_numpy()
    #_dataset_prepoc = np.concatenate( (_dataset_prepoc , cat.to_numpy()), axis=1 )
    return _dataset_prepoc, prices
print(load_data("./kc_house_data.csv"))

    
```

```python
def plot_singular_values( singulars ):
    singulars.sort()
    plt.scatter( list(range(len(singulars))), singulars[::-1])
    plt.show()

plot_singular_values(np.array([ 4 , 7 , 2 , 1 , 3, 8]))
```
